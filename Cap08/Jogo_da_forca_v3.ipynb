```python
import random
from os import system, name

def limpa_tela():
 
    if name == 'nt':
        _ = system('cls')
 
    else:
        _ = system('clear')
```

```python
def gerar_palavra():
    palavras = ['banana', 'abacate', 'uva', 'morango', 'laranja']
    palavra = random.choice(palavras)
    return palavra
```

```python
board = ['''

>>>>>>>>>>Hangman<<<<<<<<<<

+---+
|   |
    |
    |
    |
    |
=========''', '''

+---+
|   |
O   |
    |
    |
    |
=========''', '''

+---+
|   |
O   |
|   |
    |
    |
=========''', '''

 +---+
 |   |
 O   |
/|   |
     |
     |
=========''', '''

 +---+
 |   |
 O   |
/|\  |
     |
     |
=========''', '''

 +---+
 |   |
 O   |
/|\  |
/    |
     |
=========''', '''

 +---+
 |   |
 O   |
/|\  |
/ \  |
     |
=========''']
```

```python
class Hangman():
    
    def __init__(self, palavra):
        self.palavra = palavra
        self.letras_palavra = [letra for letra in palavra]
        self.letras_tentativa =  []
        self.letras_erradas = []
        self.tabuleiro = ["_"] * len(palavra)
    
    def show_status(self):
        print(board[len(self.letras_erradas)])
        print(self.tabuleiro)
        print("Letras erradas:", self.letras_erradas)
    
    def guess_letter(self):
        
        tentativa = input("\nDigite uma letra:")
        
        if tentativa in self.letras_tentativa:
            print("Você já tentou essa letra. Escolha outra!")
        
        self.letras_tentativa.append(tentativa)
        
        if tentativa in self.letras_palavra:
            print("Você acertou a letra!")
            for i in range(len(self.letras_palavra)):
                if self.letras_palavra[i] == tentativa:
                    self.tabuleiro[i] = tentativa                    
        else:
            print("Essa letra não está na palavra!")
            self.letras_erradas.append(tentativa)
        
    def isgameover(self):
        
        if "_" not in self.tabuleiro:
            self.venceu = True
            return True
        
        if len(self.letras_erradas) == 6:
            self.venceu = False
            return True   
        
        return False
    
    def isgamewon(self):
        if self.isgameover():
            if self.venceu:
                print("\nParabéns! Você venceu.")
                print("A palavra era:", self.palavra)
            else:
                print("\nGame over! Você perdeu.")
                print("A palavra era:", self.palavra)
            print("Obrigado por jogar")
    
```

```python
def game():
    
    limpa_tela()
    print("\nBem vindo ao jogo da forca!")
    print("Adivinhe a palavra abaixo:\n")
    
    palavra = gerar_palavra()
    
    game = Hangman(palavra)
    
    while not game.isgameover():
        
        game.show_status()
        
        game.guess_letter()
        
        game.isgameover()
        
    game.isgamewon()
```

```python
if __name__ == "__main__":
    game()
```
